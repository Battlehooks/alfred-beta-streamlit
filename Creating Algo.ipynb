```python
ikea_data = pd.read_csv('ikea_ds.tsv', delimiter = '\t')
ikea_data
```

```python
ikea_data = ikea_data.drop_duplicates(subset = ['name']).reset_index(drop = True)
ikea_data
```

```python
ikea_data['text'] = ikea_data['name'] + ' [SEP] ' + ikea_data['type'] + ' [SEP] ' + ikea_data['description']
ikea_data
```

```python
from langchain.document_loaders import DataFrameLoader
from langchain import HuggingFaceHub, PromptTemplate, FAISS, HuggingFacePipeline
from langchain.vectorstores import Chroma
from langchain.retrievers.multi_query import MultiQueryRetriever
from langchain.embeddings import HuggingFaceEmbeddings, OpenAIEmbeddings, SpacyEmbeddings, HuggingFaceHubEmbeddings, CacheBackedEmbeddings
from langchain.text_splitter import CharacterTextSplitter
from langchain.chat_models import ChatOpenAI
from langchain.chains import RetrievalQA
from langchain.storage import InMemoryStore, LocalFileStore, RedisStore
import os
import openai

os.environ['HUGGINGFACEHUB_API_TOKEN'] = huggingface_key
os.environ['OPENAI_API_KEY'] = openai_key
```

```python
loader = DataFrameLoader(ikea_data, page_content_column = 'imgAlt')
docs = loader.load()
```

```python
embeddings = OpenAIEmbeddings()
fs = LocalFileStore('./cache_embed/')
cache_embed = CacheBackedEmbeddings.from_bytes_store(
    embeddings, fs, namespace = embeddings.model
)
```

```python
faissdb = FAISS.from_documents(docs, cache_embed)
faissdb
```

```python
pprint(faissdb.similarity_search_with_score('Ajje'))
```
