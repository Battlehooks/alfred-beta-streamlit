```python
import numpy as np
import pandas as pd
import selenium as se
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.chrome.options import Options
from tqdm.autonotebook import tqdm
import time
import joblib
from joblib import Parallel, delayed
from pprint import pprint
tqdm.pandas()
```

```python
options = Options()
# options.add_argument('--headless')
service = ChromeService(executable_path=ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)
```

```python
ikea_data = pd.read_csv('ikea_ds.tsv', delimiter = '\t')
ikea_data
```

```python
ikea_data = ikea_data.drop_duplicates(subset = 'name', keep = 'first').reset_index(drop = True)
ikea_data
```

```python
def get_video(name, tipe) :
    driver.get(f'https://www.youtube.com/results?search_query={name}+{tipe}+assembly')
    elements = driver.find_elements(By.XPATH, '//a[@id="video-title"]')
    elements_df = pd.DataFrame(
        [[name, element.get_attribute('title'), element.get_attribute('href')] for element in elements],
        columns = ['name', 'title', 'href']
    )
    return elements_df
```

```python
all_ikea_youtube = pd.concat([
    get_video(row['name'], row['type']) for idx, row in ikea_data[['name', 'type']].drop_duplicates(subset = 'name').iterrows()
], axis = 0).reset_index(drop = True)
all_ikea_youtube
```

```python
tri_ikea_youtube = all_ikea_youtube.groupby('name').head(3).reset_index(drop = True)
tri_ikea_youtube
```

```python
ikea_data
```

```python
ikea_data = ikea_data.merge(tri_ikea_youtube, on='name').drop_duplicates().dropna()
display(ikea_data)
```

```python
url_data = ikea_data[['name', 'title', 'href']].drop_duplicates(subset = ['name', 'href'])
pure_ikea_data = ikea_data.drop(['title', 'href'], axis = 1)
```

```python
url_data.to_csv('youtube_url_ikea.tsv', sep = '\t', index = False)
pure_ikea_data.to_csv('ikea_ds.tsv', sep  = '\t', index = False)
```
